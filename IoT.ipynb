```python
import time
import json
import paho.mqtt.client as mqtt 


def on_message(client, userdata, message):
    print("Message received " ,str(message.payload.decode("utf-8")))
    print("Message topic=",message.topic)
    print("Message qos=",message.qos)
    print("Message retain flag=",message.retain)

broker_address="broker.mqttdashboard.com"

with open('myjson.json') as f:
    jsonobj = json.load(f)
    print(jsonobj)

print("New Instance Creating...")
client = mqtt.Client("mytpha4") 
for key in jsonobj:
    client.on_message=on_message 
    print("Broker Connecting....")

    client.connect(broker_address)
    client.loop_start() 

    print("Subscribing to ","temp/alert")

    client.subscribe("temp/alert")

    print("Publishing message to ","temp/alert")

    client.publish("temp/alert",key["Temperature"])

    time.sleep(4) 

    client.loop_stop() 
    client.on_message=on_message 

    print("Connecting Broker")

    client.connect(broker_address) 
    client.loop_start()

    print("Subscribing to topic","pressure/alert")

    client.subscribe("pressure/alert")

    print("Publishing message to topic","pressure/alert")

    client.publish("pressure/alert",key["Pressure"])

    time.sleep(4) 

    client.loop_stop() 
    client.on_message=on_message 
    print("Connecting Broker")
    client.connect(broker_address) 
    client.loop_start() 
    print("Subscribing to topic","humidity/alert")
    client.subscribe("humidity/alert")
    print("Publishing message to topic","humidity/alert")
    client.publish("humidity/alert",key["Humidity"])
    time.sleep(4) 
    client.loop_stop() 
    if (key["Temperature"] > 28 and key["Pressure"] > 870 and key["Humidity"] > 12):
      client.on_message=on_message 
      print("Connecting Broker..")
      client.connect(broker_address)
      client.loop_start() 
      print("Subscribing to topic","/alert")
      client.subscribe("pressure/alert")
      print("Publishing message to topic","/alert")
      client.publish("/alert","Alert")
      time.sleep(4) 
      client.loop_stop()
```

```python
import paho.mqtt.client as mqtt
import time
import json
import random

def on_message(client, userdata, message):
    print("Msg received " ,str(message.payload.decode("utf-8")))
    print("Msg topic=",message.topic)
    print("Msg qos=",message.qos)
    print("Msg retain flag=",message.retain)

broker_address="broker.mqttdashboard.com"

print("New Instance Creating...")
client = mqtt.Client("mytpha4") 
for i in range(0,10):
    temp=random.randint(25,30)
    pressure=random.randint(850,900)
    humidity=random.randint(10,15)
    client.on_message=on_message 
    print("Broker Connecting..")
    client.connect(broker_address) 
    client.loop_start() 
    print("Subscribing to topic","temp/alert")
    client.subscribe("temp/alert")
    print("Publishing message to topic","temp/alert")
    client.publish("temp/alert",temp)
    time.sleep(4) 
    client.loop_stop() 
    client.on_message=on_message
    print("Broker Connecting..")
    client.connect(broker_address) 
    client.loop_start() 
    print("Subscribing to topic","pressure/alert")
    client.subscribe("pressure/alert")
    print("Publishing message to topic","pressure/alert")
    client.publish("pressure/alert",pressure)
    time.sleep(4) 
    client.loop_stop() 
    client.on_message=on_message 
    print("Broker Connecting..")
    client.connect(broker_address) 
    client.loop_start() 
    print("Subscribing to topic","humidity/alert")
    client.subscribe("humidity/alert")
    print("Publishing message to topic","humidity/alert")
    client.publish("humidity/alert",humidity)
    time.sleep(4) 
    client.loop_stop() 
    if (temp > 28 and pressure > 870 and humidity > 12):
      client.on_message=on_message 
      print("Broker Connecting..")
      client.connect(broker_address) 
      client.loop_start() 
      print("Subscribing to topic","/alert")
      client.subscribe("pressure/alert")
      print("Publishing message to topic","/alert")
      client.publish("/alert","Alert")
      time.sleep(4) 
      client.loop_stop() 
```
